## Quantum Subspace Diagonalization (QSD)

This module implements a complete quantum computing pipeline for solving
the eigenvalue problem of molecular Hamiltonians using:

1. Krylov Subspace Construction via Trotterized Time Evolution *(NEW!)*
2. Variational Quantum Circuits for State Preparation
3. Quantum Estimator for Hamiltonian Matrix Element Measurements
4. Classical Diagonalization in the Reduced Subspace

#### THEORY BACKGROUND:

The Krylov subspace method works by creating a basis of vectors:
    $$K_n(H, |œà‚ÇÄ‚ü©) = span\{|œà‚ÇÄ‚ü©, H|œà‚ÇÄ‚ü©, H¬≤|œà‚ÇÄ‚ü©, ..., H^{n-1}|œà‚ÇÄ‚ü©\}$$

In this subspace, the Hamiltonian eigenvalue problem becomes much smaller
and easier to solve. The key insight is that low-lying eigenstates of H
can be well-approximated using linear combinations of these Krylov vectors.

*Note: A thing I forgot saying on the speak is that QSD also performs better with larger molecules, 
as the projections helps us to make this problems smaller and easier (smaller Hilbert Space == Less qubits for getting a result)*


#### WHY THIS WORKS:

- If $|œà‚ÇÄ‚ü©$ has overlap with the ground state, repeated application of $H$
  amplifies the ground state component (power method principle)
- Higher Krylov vectors span the space of excited states
- The reduced Hamiltonian $H_{eff}$ in this basis captures all the physics
  of the low-energy spectrum

#### QUANTUM ADVANTAGE:

- Classical: Building Krylov vectors requires $O(N¬≤)$ operations ($N =$ dimension)
- Quantum: Can prepare and manipulate these vectors efficiently on quantum hardware
- This implementation shows how to do this using real quantum circuits

## Build Krylov subspace using Trotterized time evolution.
    
#### WHAT IS KRYLOV SUBSPACE?
The Krylov subspace is a mathematical space spanned by vectors:
    $$|œà‚ÇÄ‚ü©, H|œà‚ÇÄ‚ü©, H¬≤|œà‚ÇÄ‚ü©, H¬≥|œà‚ÇÄ‚ü©, ... $$
    
Each vector is obtained by applying the Hamiltonian H to the previous one.
This creates a **"ladder"** of states that explore different energy levels.
    
#### HOW WE BUILD IT ON QUANTUM COMPUTER:

`Problem:` We can't directly apply H to a quantum state

`Solution:` Use time evolution $e^{-iHt} ‚âà I - iHt$, for small $t$ (Quite similar as we talk in the video of Quantum Chemistry, in the slide of VarQITE)

`Mathematical trick:`
    $$H|œà‚ü© ‚âà \frac{e^{-iHt}|œà‚ü© - |œà‚ü©}{i \cdot dt}$$

So by evolving the state for a tiny time dt and taking the difference,
we approximate $H|œà‚ü©$!

#### TROTTERIZATION:

Time evolution $e^{-iHt}$ is hard to implement directly because H is complex.
Trotterization splits it into many small steps using the Suzuki-Trotter formula:
    $$e^{-iHt} \approx \left[ e^{-iH_1 t / n} \, e^{-iH_2 t / n} \, \cdots \right]^n$$

where $H = H‚ÇÅ + H‚ÇÇ + ...$ (sum of Pauli terms)

This would be the ideal way to do it via a quantum computer (and the version I wanted to do it). Note we apply here a version of a simulation of quantum computer in a classical one. So, if we apply all of this, is going to take a few hours of our time. Therefore, in this part, we'll not use Trotterization (suzuki Trotterization) and we'll use just the matrix format (optimal value for simulation, but not quantum computing realistic)

```python
def create_evolution_circuit(H_pauli, dt, num_qubits):
    """
    Create time evolution circuit: e^(-iHt)
    Use 'matrix' synthesis for EXACT evolution.
    """
    qc = QuantumCircuit(num_qubits)
    evolution_gate = PauliEvolutionGate(H_pauli, time=dt, synthesis='matrix')
    qc.append(evolution_gate, range(num_qubits))
    
    return qc
```

```python
def qsd_projection_qiskit(H_pauli, V_krylov, name):
    """
    Perform Quantum Subspace Diagonalization (QSD) using the Krylov basis.

    Here we could perfom qiskit, but as we implemented this fully classical
    via Statevectors and matrix PauliEvolutionGates, let's keep this format
    so you can see how can we apply a QSD via matrix implementation.
    
    ================================================================
    Theoretical Overview
    ================================================================
    Once we have a reduced Krylov subspace spanned by the orthonormal 
    columns of V, we can project the large Hamiltonian H into this 
    smaller subspace:
    
        H_eff = V‚Ä† H V
    
    This gives an effective Hamiltonian H_eff that captures the 
    low-energy behavior of the full system but is much smaller.
    
    We then diagonalize H_eff to obtain approximate eigenvalues and 
    eigenvectors ‚Äî this is the heart of QSD.
    
    Parameters
    ----------
    H_pauli : Operator
        Original system Hamiltonian.
    V_krylov : np.ndarray
        Matrix with orthonormal basis vectors as columns.
    name : str
        Identifier for printing/logging.
    
    Returns
    -------
    evals : np.ndarray
        Eigenvalues (energies) of the effective Hamiltonian.
    evecs_full : np.ndarray
        Approximate eigenvectors in the full Hilbert space.
    """
    
    print(f"\nüî¨ QSD Projection: {name}")
    
    dim = V_krylov.shape[1]
    print(f"   Subspace dimension: {dim}")
    
    # ------------------------------------------------------------
    # 1. GET FULL HAMILTONIAN MATRIX
    # ------------------------------------------------------------
    H_matrix = H_pauli.to_matrix()
    
    # ------------------------------------------------------------
    # 2. PROJECT INTO SUBSPACE
    # ------------------------------------------------------------
    H_eff = V_krylov.conj().T @ H_matrix @ V_krylov
    
    # Ensure Hermiticity (symmetrize numerically)
    H_eff = (H_eff + H_eff.conj().T) / 2
    
    # ------------------------------------------------------------
    # 3. DIAGONALIZE THE EFFECTIVE HAMILTONIAN
    # ------------------------------------------------------------
    evals, evecs_sub = np.linalg.eigh(H_eff)
    
    print(f"\n   Eigenvalues (lowest 5):")
    for i, E in enumerate(evals[:min(5, len(evals))]):
        print(f"   E_{i} = {E:.8f} Ha")
    
    # ------------------------------------------------------------
    # 4. RECONSTRUCT FULL EIGENVECTORS
    # ------------------------------------------------------------
    evecs_full = V_krylov @ evecs_sub
    
    for i in range(min(3, evecs_full.shape[1])):
        norm = np.linalg.norm(evecs_full[:, i])
        if abs(norm - 1.0) > 1e-6:
            print(f"   ‚ö†Ô∏è  Warning: |œÜ_{i}‚ü© norm = {norm:.6f}")
    
    print(f"   ‚úÖ QSD complete!\n")
    
    return evals, evecs_full
```

```python
def create_varqite_ansatz(n_qubits, depth=3):
    """
    Some manual ansatz for  VarQITE
     RY + RZ + circular entanglement 
    """
    n_params = n_qubits * 2 * depth + n_qubits
    params = ParameterVector('Œ∏', n_params)
    
    qc = QuantumCircuit(n_qubits)
    param_idx = 0
    
    for layer in range(depth):
        # Rotaciones RY + RZ
        for q in range(n_qubits):
            qc.ry(params[param_idx], q)
            param_idx += 1
            qc.rz(params[param_idx], q)
            param_idx += 1
        
        # Entanglement circular
        for q in range(n_qubits - 1):
            qc.cx(q, q + 1)
        if n_qubits > 2:
            qc.cx(n_qubits - 1, 0)
    
    # Capa final de rotaciones
    for q in range(n_qubits):
        qc.ry(params[param_idx], q)
        param_idx += 1
    
    return qc, params
```
